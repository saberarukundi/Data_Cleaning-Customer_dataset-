#### apply() => Apply is a pandas  function  which will apply specified functionality to each value of series 

```python
df['Last_Name'] = df['Last_Name'].str.strip("123._/")
```

```python
df['Last_Name']
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x:str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace("[^0-9]","",regex=True)
```

```python
l1=lambda x:x[0:3]+ "-"+ x[3:6] + "-" + x[6:10]
```

```python
df['Phone_Number']=df['Phone_Number'].apply(l1)
```

```python
df.replace('--','',inplace=True)
```

```python
import numpy as np

# N/a, None = >null values
type('N/a')
```

```python
df.fillna('',inplace=True)
```

```python
for i in df.index:
    if df.loc[i,'Do_Not_Contact'] == "Y":
        df.drop(i,inplace=True)
```
